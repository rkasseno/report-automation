```python
df = df.map(lambda x: " ".join(x.split()) if isinstance(x, str) else x)
```

```python
table = df.groupby(['Student_IIN'])['Student_Name'].value_counts().reset_index()
table
```

```python
iins = {}
duplicate_iins = set()

for iin, name in zip(table['Student_IIN'], table['Student_Name']):
    if not iin in iins:
        iins[iin] = name
    else:
        if not iins[iin] == name:
            duplicate_iins.add(iin)
```

```python
print(duplicate_iins)
len(duplicate_iins)
```

```python
t = df[df['Student_IIN'].isin(duplicate_iins)][['Student_IIN', 'Student_Name']].assign(iin_number = lambda x: x.groupby(['Student_IIN']).ngroup()).sort_values(by='iin_number')
t_grouped = t.groupby('Student_IIN')['Student_Name'].value_counts().reset_index()
t_grouped
```
