```python
import numpy as np
cols = np.arange(1, 21)
cols = [str(col) for col in cols]

df[cols] = df['answers'].apply(lambda x: pd.Series(list(x))).astype(int)
df[cols] = df[cols].map(lambda x: x - 9 if x == 9 else x)
```

```python
to_melt = df.drop(columns=['school', 'school_id', 'lang', 'lang_test', 'iin', 'name', 'answers', 'score'])
```

```python
melted_df = to_melt.melt(id_vars=['grade', 'subject'], value_vars=cols).rename(columns={'variable' : 'question', 'value' : 'answer'})
```

```python
melted_grouped = melted_df.groupby(['grade', 'subject', 'question'])['answer'].sum().reset_index()
melted_grouped
```

```python
melted_count = melted_df.groupby(['grade', 'subject'])['question'].value_counts().reset_index()
melted_count
```

```python
melted_merged = pd.merge(melted_grouped, melted_count, on=['grade', 'subject', 'question'])
melted_merged = melted_merged.assign(percent = lambda x: x['answer'] / x['count'] * 100).round(1)
melted_merged
```

```python
elem_df = pd.read_excel(r"C:\Users\kassenov_r\OneDrive - AEO NIS\Рабочий стол\Мониторинг\Сельские школы\data\elements.xlsx")
elem_df = elem_df.map(lambda x: " ".join(x.split()) if isinstance(x, str) else x)
```

```python
elem_df = elem_df.assign(grade = lambda x: x['subject'].apply(lambda z: 8 if '8' in z else 4))
elem_df['subject'] = elem_df['subject'].str[1:]
```

```python
subject_map = {
    'PHYS' : 'Физика', 
    'BIOL' : 'Биология', 
    'ENFL' : 'Английский язык', 
    'SCIE' : 'Естествознание и познание мира', 
    'MATH' : 'Математика', 
    'MATH' : 'Математика', 
    'RUFL' : 'Русский язык (Я1)',
    'RUFL' : 'Русский язык (Я1)',
    'RUSL' : 'Русский язык (Я2)', 
    'CHEM' : 'Химия', 
    'KZFL' : 'Казахский язык (Я1)', 
    'KZFL' : 'Казахский язык (Я1)', 
    'KZSL' : 'Казахский язык (Я2)'
}
elem_df['subject'] = elem_df['subject'].map(subject_map)
```

```python
elem_df
```

```python
melted_merged['question'] = melted_merged['question'].astype(int)
elem_merged = pd.merge(elem_df, melted_merged, on=['subject', 'grade', 'question'], how='left')
```

```python
elem_merged[elem_merged['subject'] == 'Английский язык']
```

```python
elem_merged['grade'] = elem_merged['grade'].astype(str)
grades = [str(grade) for grade in grades]
file_path = r"C:\Users\kassenov_r\OneDrive - AEO NIS\Рабочий стол\Мониторинг\Сельские школы\presentation\overall\excel_new\element_tables.xlsx"
with pd.ExcelWriter(file_path) as writer:
    for grade in grades:
        for subject in subjects:
            table = elem_merged[(elem_merged['grade'] == grade) & (elem_merged['subject'] == subject)][['question', 'element', 'percent']].sort_values(by='percent', ascending=False)
            if len(table) > 0:
                table.to_excel(writer, sheet_name=f"{grade}_{subject}", index=False)
```
